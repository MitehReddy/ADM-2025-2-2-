```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/titanic.csv')
```

```python
print("Updated Statistical Summary:")
print(data.describe())
print("\nUpdated General Info:")
print(data.info())
```

```python
plt.figure(figsize=(12, 7))
sns.histplot(data['Age'].dropna(), kde=True, bins=30, color='blue')
plt.title('Age Distribution of Passengers', fontsize=18)
plt.xlabel('Age (Years)', fontsize=16)
plt.ylabel('Frequency', fontsize=16)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 7))
sns.countplot(x='Pclass', data=data, palette='coolwarm')
plt.title('Passenger Class Distribution', fontsize=18)
plt.xlabel('Passenger Class', fontsize=16)
plt.ylabel('Count of Passengers', fontsize=16)
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(12, 7))
sns.boxplot(x='Pclass', y='Age', data=data, palette='viridis')
plt.title('Age Distribution Across Passenger Classes', fontsize=18)
plt.xlabel('Passenger Class', fontsize=16)
plt.ylabel('Age (Years)', fontsize=16)
plt.show()
```

```python
sns.pairplot(data[['Age', 'Fare', 'SibSp', 'Parch']].dropna(), diag_kind='hist', corner=True, palette='flare')
plt.suptitle('Pairwise Relationship of Features', y=1.02, fontsize=18)
plt.show()
```

```python
plt.figure(figsize=(14, 9))
correlation = data[['Age', 'Fare', 'SibSp', 'Parch']].corr()
sns.heatmap(correlation, annot=True, cmap='magma', fmt='.2f', linewidths=0.5)
plt.title('Feature Correlation Heatmap', fontsize=18)
plt.show()
```

```python
plt.figure(figsize=(12, 7))
sns.violinplot(x='Pclass', y='Fare', data=data, palette='rocket', scale='width', inner='quartile')
plt.title('Fare Distribution Across Passenger Classes', fontsize=18)
plt.xlabel('Passenger Class', fontsize=16)
plt.ylabel('Fare Amount ($)', fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(12, 7))
sns.countplot(x='Embarked', data=data, palette='pastel', order=data['Embarked'].value_counts().index)
plt.title('Passenger Count by Embarkation Port', fontsize=18)
plt.xlabel('Embarkation Port', fontsize=16)
plt.ylabel('Number of Passengers', fontsize=16)
plt.grid(axis='y')
plt.show()
```

```python
data['AgeGroup'] = pd.cut(data['Age'], bins=range(0, 101, 10), labels=['0-9', '10-19', '20-29', '30-39', '40-49', '50-59', '60-69', '70-79', '80-89', '90-99'])
avg_fare_by_age = data.groupby('AgeGroup')['Fare'].mean().reset_index()

plt.figure(figsize=(14, 9))
sns.lineplot(x='AgeGroup', y='Fare', data=avg_fare_by_age, marker='s', color='crimson', linewidth=2.5)
plt.title('Average Fare Paid by Age Group', fontsize=18)
plt.xlabel('Age Group (Years)', fontsize=16)
plt.ylabel('Average Fare ($)', fontsize=16)
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
